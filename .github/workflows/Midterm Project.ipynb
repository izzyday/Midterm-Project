```python
import pandas as pd
numbers_file = open('/Users/izzyday/Downloads/midterm-project/numbers.txt', 'r')

file_lines = numbers_file.readlines()
file_lines = [line[:-1] for line in file_lines]

table = [line.split(',') for line in file_lines]

#I made a list of all the items in the file without line breaks
fulllist = []
for list in table:
    for value in list:
        #there was a stray space in the file, so I check to see if it is a number
        if value.isdigit():
            fulllist.append(int(value))
        
counts = pd.Series(fulllist).value_counts().to_dict()
print(counts)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = fulllist
plt.hist(x)
plt.show()  
```

```python
import pandas as pd
#read the json file into a dataframe
df = pd.read_json('~/Downloads/posts/your_posts_1.json')

df.head(3)
```

```python
#rename the timestamp column
df.rename(columns={'timestamp':'date'}, inplace=True)
#drop some unecessary columns
df = df.drop(['attachments', 'title','tags'], axis=1)
#make sure it's datatime format
pd.to_datetime(df['date'])

df.head(3)
```

```python
count=[]
for i in range(df.shape[0]):
    if len(df.iloc[i,1]) !=0:
        count.append(len(str(list(df.iloc[i,1][0].values())[0])))
    else:
        count.append(0)
df['letter_number']=count
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#set figure size and font size
sns.set(rc={'figure.figsize':(40, 20)})
sns.set(font_scale=3)
#set x labels
x_labels = post_counts.index
#create bar plot
sns.barplot(x_labels, post_counts, color="blue")
#only show x axis labels for Jan 1 of every other year
tick_positions = np.arange(10, len(x_labels), step=24)
#reformat date to display year only plt.ylabel("post counts")
plt.xticks(tick_positions, x_labels[tick_positions].strftime("%Y"))
plt.show()
```
